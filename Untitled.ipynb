```python
import concurrent.futures as futures
from collections import OrderedDict
import requests

id_to_name = resultdict
print(id_to_name)

resultdict=OrderedDict()
def load_url(url_no, timeout):
    retry = 3
    url = "http://m.stock.naver.com/api/item/getPriceDayList.nhn?code={:0=6d}&pageSize={}".format(int(url_no),999999)
    print(url_no)
    for _ in range(retry):
        res = requests.get(url, timeout=timeout)
        if res.status_code == 200:
            rows = res.json()['result']['list']
            return rows
        #print("retry {} ".format(_),end=" , ")
    print("### retried 3 times, but fail : {}".format(url))
    return

with futures.ThreadPoolExecutor(max_workers=10) as exe:
    future_to_url = {exe.submit(load_url, url,2):url for url in id_to_name}
    for key in futures.as_completed(future_to_url):
        url = future_to_url[key]
        try :
            data = key.result()
        except Exception as e:
            print(e)
        else:
            if data is not None:
                resultdict[url]= data
    print("\n\n{} : {}".format(url,data))
```
